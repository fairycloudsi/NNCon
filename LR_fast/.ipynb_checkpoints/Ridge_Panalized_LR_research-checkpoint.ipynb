# Ridge Penalized Logistic Regression $1d$

```python
def LR_function_1d(x,a,b,labd):
    return np.log(1+np.exp(np.dot(x,a))) -b*np.dot(x,a) + labd*np.square(x)
```

## $\lambda = 1$

### $b = 1$ 

```python
a = -0.5
b1 = 1.
b2 = 0.
labd = 1
```

```python
x = np.arange(-2., 2., 0.01)
```

```python
y = LR_function_1d(x,a,b1,labd)
y2 = LR_function_1d(x,a,b2,labd)
```

```python
print(y[0],y[-1])
```

```python
print(y2[0], y2[-1])
```

```python
plt.plot(x,y,'r-', x,y2,'b-')
plt.xlabel('x')
plt.ylabel('RPLR cost')
plt.title('PRLR cost function over x')
save_fig('PRLR_cost_function_a'+str(a)+'labd'+str(labd))
plt.show()
```

Conclusion:

```python
## Generate A 2*2 matrix and b 
np.random.seed(1024)
d = 2
cor_X = 0
S = cor_X * np.ones((d,d)) + (1 - cor_X)*np.diag(np.ones(d))
R = np.linalg.cholesky(S)

A = np.dot(np.random.normal(size= 2*2).reshape(2,2),R)
A = preprocessing.scale(A) * np.sqrt(float(d - 1)/d)
```
